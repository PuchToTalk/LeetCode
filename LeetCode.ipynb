### 1920. Build Array from Permutation (89,8%)

Capture d’écran 2023-07-04 à 11.24.41.png

<font color='cyan'> <ins> Idée principale </ins>: </color>
- Définir une nouvelle liste vide: "ANS"
- Boucler sur (0 , la taille de l'array - 1):
  - ajouter dans la nouvelle liste "ANS" la valeur associée à la modification souhaitée
- Retourner la nouvelle liste "ANS" comme voulue


```python
   def buildArray(nums):
        ans = [] # définit nouvelle liste
        print(f"la liste actuelle donne : {nums}")
        for i in range(len(nums)): # boucle sur la taille de la liste
          print(f"la valeur de la nouvelle indice : nums[i] donne : {nums[i]} et la nouvelle valeur associée à cet indice donne {nums[nums[i]]}")
          ans.append(nums[nums[i]]) # ajoute la nouvelle valeur à la nouvelle liste
        return ans
```

```python
nums = [0,2,1,5,3,4]


buildArray(nums)
```

### 14. Longest Common Prefix (41,1%)

<font color='cyan'> <ins>Idée principale :</ins>
- Prendre l'élément la plus petite taille (avec méthode min) : "SHORTEST"
- Itérer sur chaque lettre du mot de plus petite taille : avec " **i** " : désigne <ins> l'indice</ins> , " **char** " désigne la <ins> lettre correspondante </ins>
  - Boucler à nouveau, sur chaque élément de "STRS" (3 éléments car 3 mots)
    - Comparaison : Si à l'indice "i" la lettre de "string" est différent que la lettre du "SHORTEST" alors afficher l'élément "short" du début jusqu'à cet indice
- Enfin afficher "SHORTEST"

----

<ins>__La fonction enumerate()__ :</ins>
The enumerate() function is commonly used when you need to **iterate over elements of a sequence** while **keeping track of their indices simultaneously**.

<ins>__Explanation__ :</ins> The loop iterates over each character in the shortest string. At each iteration, **" i " holds the index of the current character**, and **" char " holds the value of the character itself**. The code inside the loop can then perform operations or comparisons based on the index and value of each character.



```python
def longestCommonPrefix(strs) -> str:
        # Find the shortest string in the array
        shortest = min(strs, key=len)
        for i, char in enumerate(shortest):
          print(f"la valeur de i donne {i}, et celle de char donne {char}")
          for string in strs:
            if string[i] != char:
              return shortest[:i]
        return shortest
```

```python
strs = ["flower","flow","flight"]


longestCommonPrefix(strs)
```
