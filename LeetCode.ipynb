----

> <ins>__La fonction enumerate()__ :</ins>
The enumerate() function is commonly used when you need to **iterate over elements of a sequence** while **keeping track of their indices simultaneously**.

<ins>__Explanation__ :</ins> The loop iterates over each character in the shortest string. At each iteration, **" i " holds the index of the current character**, and **" char " holds the value of the character itself**. The code inside the loop can then perform operations or comparisons based on the index and value of each character.



```python
def longestCommonPrefix(strs) -> str:
        # Find the shortest string in the array
        shortest = min(strs, key=len)
        for i, char in enumerate(shortest):
          print(f"la valeur de i donne {i}, et celle de char donne {char}")
          for string in strs:
            if string[i] != char:
              return shortest[:i]
        return shortest
```

```python
strs = ["flower","flow","flight"]


longestCommonPrefix(strs)
```

### 9. Palindrome Number

Capture d’écran 2023-07-05 à 16.33.43.png

<font color='cyan'> <ins>Idée principale :<ins>
- Transformer en string l'élément d'entrée
- Renverser le string dans l'autre sens
- Test booléen si on a exactement la même valeur entre valeur initiale et valeur inversée


```python
def isPalindrome(x):
        # Convertir l'entier en chaîne de caractères
        x_str = str(x)
        # Inverser la chaîne de caractères
        reversed_str = x_str[::-1]
           # Convertir la chaîne de caractères inversée en entier
        res = reversed_str == str(x)

        return res
```

<font color='cyan'> <ins>Idée principale :<ins>
- Premier Test pour tous les éléments négatifs : les balayer car tous impossible palindrome
- Puis, Déterminer le nbre de digits total dans x


```python
def isPalindrome(x):
    # Handle negative numbers (they can't be palindromes)
    if x < 0:
        return False

    # Determine the number of digits in x
    num_digits = 0
    temp = x
    while temp > 0:
        temp //= 10
        num_digits += 1

    # Compare digits from both ends of the number
    left = x
    right = 0
    for _ in range(num_digits // 2):
        right = right * 10 + left % 10
        left //= 10
        print(f"Valeur de right donne {right} \nValeur de left donne {left}")

    # If the number of digits is odd, ignore the middle digit
    if num_digits % 2 != 0:
        left //= 10

    # Check if the left and right halves are equal
    return left == right
```

<font color='red'> <ins>Explication :</ins> </font>
Dans le code fourni, la ligne right = right * 10 + left % 10 est utilisée pour construire l'inverse de la moitié droite du nombre.

Voici comment cela fonctionne dans le contexte du code :

right = right * 10 : À chaque itération de la boucle for, right est multiplié par 10. Cela est fait pour déplacer les chiffres existants dans right vers la gauche, créant ainsi de l'espace pour le nouveau chiffre à ajouter.

left % 10 : L'expression left % 10 calcule le chiffre le plus à droite de left. L'opérateur de modulo % renvoie le reste de la division de left par 10, extrayant ainsi le chiffre le plus à droite.

right = right * 10 + left % 10 : Cette instruction combine la valeur précédente de right, déplacée vers la gauche d'un chiffre (right * 10), avec le chiffre le plus à droite de left (left % 10). Cela permet de construire l'inverse de la moitié droite du nombre en ajoutant chaque chiffre suivant de left à droite de right.

L'objectif de cette opération est de créer une version inversée de la moitié droite du nombre afin de pouvoir la comparer avec la moitié gauche pour déterminer si le nombre est un palindrome.

L'instruction print que vous avez ajoutée est utile pour comprendre les valeurs intermédiaires de right et left à chaque itération de la boucle. Elle peut être utilisée pour le débogage ou pour observer le comportement du code.

```python
x = 121


isPalindrome(x)
```

```python
x = -121

isPalindrome(x)
```

```python
x = 10

isPalindrome(x)
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

    def minDepth(self, root):
        if not root:
            return 0
        elif not root.right and not root.left:
            return 1
        elif not root.right:
            return 1 + self.minDepth(root.left)
        elif not root.left:
            return 1 + self.minDepth(root.right)
        return min(self.minDepth(root.left), self.minDepth(root.right)) + 1


# Create the binary tree
root = TreeNode(3)
root.left = TreeNode(9)
root.right = TreeNode(20)
root.right.left = TreeNode(15)
root.right.right = TreeNode(7)

# Calculate the minimum depth
result = root.minDepth(root)
print(result)
```

" If my_list.count(target_element) > 0 ", it means that the *target_element* appears at least once in the my_list.
So we may put > 1, to see if there is more than double appearance.

> **The count() method** returns the number of occurrences of the *target_element* in the *my_list*. By checking if the count is greater than 0, we can determine if the element exists in the list.

```python
 def singleNumber(nums):
        res = []
        repet = []
        for i in range(len(nums)-1, -1, -1):
            if nums.count(nums[i]) > 1:
              print(f"la valeur à indice i = {i} est de {nums[i]}")
              repet.append(nums[i])
              nums.remove(nums[i])
            else:
              if nums[i] in repet:
                nums.remove(nums[i])
              else:
                extracted_element = nums[i]
                res.append(extracted_element)
                print(res)

        return extracted_element
```

<ins>__Simplification du code__:</ins>
- On rassemble les deux parties nums.remove ensemble grâce au OR
- Ainsi on récupère directement la SINGLE VALUE "*extracted_element*", on obtient le résultat souhaité


```python
 def singleNumber(nums):
  repet = []
  for i in range(len(nums)-1, -1, -1):
    if nums.count(nums[i]) > 1 or nums[i] in repet:
      repet.append(nums[i])
      nums.remove(nums[i])

    else:
      extracted_element = nums[i]
  return extracted_element
```

```python
nums1 = [2,2,3,2]
nums2 = [0,1,0,1,0,1,99]


print(f"La valeur unique de la liste {nums1} donne : {singleNumber(nums1)}. \nPuis la valeur unique de la liste {nums2} donne : {singleNumber(nums2)}")
```
